```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_excel("TRAIN.xlsx")
```

```python
nan_cols = [i for i in df.columns if df[i].isnull().all()]
df=df.drop(nan_cols, axis=1)
```

```python
nan = df[df.columns[df.isnull().mean() >= 0.8]]
df.drop(nan, axis=1, inplace=True)
```

```python
df = df.fillna(0)
```

```python
df['PREFERRED_LANGUAGE'] = df['PREFERRED_LANGUAGE'].replace('KAZ',1)
df['PREFERRED_LANGUAGE'] = df['PREFERRED_LANGUAGE'].replace('RUS',2)
df['PREFERRED_LANGUAGE'].value_counts()
```

```python
df['GENDER'] = df['GENDER'].apply(lambda x: 1 if x == "F" else 0)
df
```

```python
new = df['DATE_FRM_LST_KASPIKZ_PMT']
new
```

```python
from datetime import datetime
now = datetime.now()
type(now)
```

```python
aa=[]
for i in new:
    if i!=0:
        a = now-datetime.strptime(i, '%d%b%Y:%H:%M:%S')
        aa.append(a.days)
    else:
        aa.append(0)
df['DATE_FRM_LST_KASPIKZ_PMT'] = aa
```

```python
df['DATE_FRM_LST_KASPIKZ_PMT'] = aa
```

```python
df = df.drop(['KDN_GCVP','KDN_ANKETA','KDN_SUMMARY'], axis=1)
```

```python
df_x = df.drop(['TARGET'], axis=1)
```

```python
df_y = df['TARGET']
```

```python
objcol = df_x.select_dtypes(include=['object'])
df_x = df_x.drop(objcol, axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, train_size=0.2)
```

```python
model1 = LogisticRegression(solver="lbfgs")
model1.fit(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```
